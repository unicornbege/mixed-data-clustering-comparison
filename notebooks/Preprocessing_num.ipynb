F. cek outliers pada data

```python
import numpy as np

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers

numeric_students = students_subset.select_dtypes(include=np.number).columns


for col in numeric_students:
    out = detect_outliers_iqr(students_subset, col)
    print(f"{col}: {len(out)} outliers")


```

pada output tersebut mengetahui adanya 1 outliers pada atribut Avg_Daily_Usage_Hours

G. cek nilai outliers pada atribut Avg_Daily_Usage_Hours

```python
# Get the outliers for Avg_Daily_Usage_Hours
outliers = detect_outliers_iqr(students_subset, 'Avg_Daily_Usage_Hours')

# Display the outliers with relevant columns
print("Outliers in Avg_Daily_Usage_Hours:")
print(outliers[['Avg_Daily_Usage_Hours', 'Age', 'Gender', 'Academic_Level', 'Country']])

# Calculate the IQR values for reference
Q1 = students_subset['Avg_Daily_Usage_Hours'].quantile(0.25)
Q3 = students_subset['Avg_Daily_Usage_Hours'].quantile(0.75)
IQR = Q3 - Q1

print(f"\nSummary Statistics:")
print(f"Q1: {Q1:.2f}")
print(f"Q3: {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Lower bound: {(Q1 - 1.5 * IQR):.2f}")
print(f"Upper bound: {(Q3 + 1.5 * IQR):.2f}")
```
