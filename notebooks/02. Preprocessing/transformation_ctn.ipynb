Melakukan perubahan nilai data ketegorik menjadi numerik menggunakan ordinalencoder pada dataset students

```python
# Baca dataset
df_students = pd.read_csv("/Users/althaaframadhan/Documents/Skripsi/penelitian/preprocessing/students_200.csv")

# Bersihkan nama kolom
df_students.columns = df_students.columns.str.strip()

# Pilih kolom kategorik (object/category)
cat_cols = df_students.select_dtypes(include=['object', 'category']).columns.tolist()
print("Kolom kategorik:", cat_cols)

# Inisialisasi encoder
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

# Fit & transform hanya kolom kategorik
df_students[cat_cols] = encoder.fit_transform(df_students[cat_cols])

# Convert ke Int64 agar tidak ada 1.0 menjadi 1.0 (float)
for col in cat_cols:
    df_students[col] = df_students[col].astype('Int64')

# Cek hasil transformasi
print(df_students.head())
print("\nTipe data setelah transformasi:")
print(df_students.dtypes)
```

```python
df_students.to_csv("/Users/althaaframadhan/Documents/Skripsi/penelitian/preprocessing/students_200", index=False)
for col, cats in zip(cat_cols, encoder.categories_):
    mapping = {cat: i for i, cat in enumerate(cats)}
    print(f"{col}: {mapping}")
```

Pada output code diatas menampilkan nilai nilai yantelah diubah dari kategorik menjadi numerik, setiap nulai numerik mmeberikan inisialiosasi pada nilai kategorik
