F. cek outliers pada data

```python
import numpy as np

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers

numeric_trends = trends_subset.select_dtypes(include=np.number).columns


for col in numeric_trends:
    out = detect_outliers_iqr(trends_subset, col)
    print(f"{col}: {len(out)} outliers")
```

output memberikan bahwa tidak ada outlier dari 200 dataset trends tersebut
